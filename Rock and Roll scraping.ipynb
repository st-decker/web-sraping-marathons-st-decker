```python
# Create a function that reads the table information
def table_info(race_results, url):
    entire_table = []
#     iterate through the table and look at the rows
    for tr in race_results.findAll('tr'):
#         in the rows, grab the cell info (th)
        table_cell = [url]
        for td in tr.findAll('td'):
            table_cell.append(td.text)
        entire_table.append(table_cell)
    return entire_table[1:]
```

```python
def find_next(soup):
    return soup.find('div',{'class':'pagination-container'}) \
            .findAll('li', {'class': 'PagedList-skipToNext'})[-1] \
            .find('a').get('href', None)
    
```

```python
# Now that we've created the soup.. Let's find the pagination elements.
# Check the repositories page to find out what to look for!
final_data = []
for race_page in MARATHONS_CONFIG['pages']:
    pageNum = 1
#     final_data = []
    start_url =MARATHONS_CONFIG['base_url'] + '/' + race_page + '?' + MARATHONS_CONFIG['query_params'] \
                              + str(pageNum)
    
#     soup = BeautifulSoup(resp.text, 'html.parser')
    
#      race_results = soup.find('table', {'class':'table'})
    next_url = start_url
    while next_url:
        print(next_url)
        resp = requests.post(next_url)
        print(resp.status_code)
        soup = BeautifulSoup(resp.text, 'html.parser')
        next_url = find_next(soup)
        rows = table_info(soup.find('table', {'class':'table'}), next_url)
        final_data.append(rows)
#     time.sleep(2)
#   final_data.append(rows)
```
